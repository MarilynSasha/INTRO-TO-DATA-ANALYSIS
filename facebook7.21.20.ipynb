```python
# x = 'type'
# y = 'like'
sns.set(rc={'figure.figsize': (10, 5)})
 
sns.set_style('ticks')
ax = sns.boxplot(x= 'like', y = 'Type', data = facebook)
```

```python
ax = sns.barplot(x= 'Type', y = 'like', data = facebook)
```

```python
ax = sns.swarmplot(x='Post Month', y="like", data=facebook, hue = 'Type')
```

```python
ax = sns.scatterplot(x='Post Month', y="like", data=facebook, hue = 'Type')
```

```python
sns.set(rc={'figure.figsize': (12, 6)})

ax = sns.lineplot(x = 'Post Month', y = 'like', data = facebook, style = 'Type', hue = 'Type', 
                  palette = ['purple', 'green', 'blue', 'orange'],  ci = False)

ax.set_xticks(range(1, 13, 1)) 



# from christian: plt.xticks(np.arange(0, 100, 5))
# ax.set_xticks(range(1, 13, 1)) 
```
